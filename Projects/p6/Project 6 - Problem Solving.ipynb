A "Caesar code" uses an alphabet treated like a circle and a number key from 1 through 25.<br><br>

```

```

````python
# 1)  4 points Caesar code

#     We've worked on a Caesar code in class,
#     Encode a message using a key of 6
#     Decode the following message using a key of 12


def encode_letter(letter, key):
    ''' Given a lower-case letter and a number key like 3 or -2, return the Caesar Code letter '''
    
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    length = len(alphabet)
    if letter not in alphabet:
        return letter
    
    number = ord(letter) - ord('a')
    code_number = number + key
    code_position = code_number % length  # remainder
    code_letter = alphabet[code_position]
    return code_letter


def decode_letter(letter, key):
    ''' Given a lower-case letter and a key, return the decoded letter '''
    
    my_key = - key
    good_letter = encode_letter(letter, my_key)
    return good_letter



# Encoding my message
msg = 'i finally got it dad!'
code_msg = ''
key = 6
for character in msg:
    code = encode_letter(character, key)
    code_msg = code_msg + code

print("My coded message:", code_msg)
print()


# Original part of the project
# --------------------------------------------------------------------
secret = "qhqdkftuzs etagxp nq me euybxq me baeeunxq, ngf za euybxqd."
key = 12
# --------------------------------------------------------------------

# For decoding the secret
good_secret = ''
for char in secret:
    good_letter = decode_letter(char, key)
    good_secret = good_secret + good_letter
    
print("The decoded given secret:", good_secret)
````

````python
# 8)  5 pts

# You can assume your list has at least 2 distinct numbers in it.

# You could test your program using these two lists

first_list = [5, 6, 1, 16, 87, 3, 4, 2, 15] # Should get 1, 2

another_list = [65, 87, 119, 21, 55, 21, 32, 97] # Should get 21, 32

final_list = [3, 4, 6]

# Your code here:


def second_smallest(number_list):
    n1, n2 = float('inf'), float('inf')
    for x in number_list:
        if x < n1:
            n1, n2 = x, n1
        elif x < n2:
            n2 = x
    return n2

print("The two smallest numbers in the first list are:\t", min(first_list), second_smallest(first_list))
print("The two smallest numbers in another list are:\t", min(another_list), second_smallest(another_list))
print("The two smallest numbers in the first list are:\t", min(final_list), second_smallest(final_list))
````
